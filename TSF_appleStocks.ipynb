# Time Series Forecasting for Apple stocks prices

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
```

### 1. Connect to postgres database

```python
engine = create_engine('postgresql+psycopg2://postgres:chechoreyesdev@127.0.0.1:5432/postgres')
# engine = create_engine("postgresql:///?User=postgres&;Password=chechoreyesdev&Database=postgres&Server=127.0.0.1&Port=5432")
```

```python
df = pd.read_sql("SELECT * FROM stocks_apple ORDER BY stock_id ASC ", engine, parse_dates=['Date'])
```

```python
df.head()
```

```python
df1=df.loc[:,['Date','Close']]
df1.set_index('Date',inplace=True)
df1.plot(figsize=(12,5))
plt.ylabel('USDollar')
plt.legend().set_visible(False)
plt.tight_layout()
plt.title('Apple stock prices')
sns.despine(top=True)
plt.show()
```
